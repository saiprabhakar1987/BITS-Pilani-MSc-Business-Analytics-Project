```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Load the dataset
df = pd.read_excel('Barley_Prices_Aggregated.xlsx')
```

```python
# Split data into training and testing sets
train, test = train_test_split(df, test_size=0.2)
```

```python
# Extract features and target variable
train_X = train.drop(columns=['Modal Price_Barley'])
train_y = train['Modal Price_Barley']
test_X = test.drop(columns=['Modal Price_Barley'])
test_y = test['Modal Price_Barley']
```

```python
# Initialize the Gradient Boosting model
gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
```

```python
# Fit the model on the training data
gb_model.fit(train_X, train_y)
```

```python
# Make predictions on the testing data
predictions = gb_model.predict(test_X)
```

```python
# Calculate the Mean Squared Error
mse = mean_squared_error(test_y, predictions)
print("Mean Squared Error: %.2f" % mse)
```

```python
rmse = np.sqrt(mse)
```

```python
print('RMSE: {:.2f}'.format(rmse))
```

```python
# Plot the actual and predicted values
plt.title('GBM - Barley Price Prediction')
plt.plot(test_y.values, label='Actual')
plt.plot(predictions, label='Predicted')
plt.legend()
plt.show()
```
